```python
import os
import subprocess
```

```python
# Base paths for raw and processed datasets
raw_base_path = "/content/drive/Shareddrives/ML Project/Dataset/Raw"
processed_base_path = "/content/drive/Shareddrives/ML Project/Dataset/Processed"

# Ensure the processed base directory exists
if not os.path.exists(processed_base_path):
    os.makedirs(processed_base_path)
```

```python
# Function to process videos
def process_videos(raw_dir, processed_dir):
    for root, dirs, files in os.walk(raw_dir):
        for dir in dirs:
            # Create a corresponding directory in the processed path
            raw_path = os.path.join(root, dir)
            processed_path = os.path.join(root.replace(raw_dir, processed_dir), dir)
            if not os.path.exists(processed_path):
                os.makedirs(processed_path)

        for file in files:
            if file.lower().endswith(('.mov', '.mp4')):
                full_raw_path = os.path.join(root, file)
                full_processed_path = os.path.join(root.replace(raw_dir, processed_dir), f"{file}")
                # ffmpeg command to cut the first 60 frames and resize the video
                command = [
                    "ffmpeg",
                    "-i", full_raw_path,       # Input file
                    "-vf", "crop=min(iw\,ih):min(iw\,ih),scale=540:540",    # Resize filter
                    "-frames:v", "60",         # Extract 60 frames
                    "-an",                     # Remove audio
                    full_processed_path        # Output file
                ]
                try:
                    subprocess.run(command, check=True)
                    print(f"Processed {full_raw_path} successfully.")
                except subprocess.CalledProcessError as e:
                    print(f"Failed to process {full_raw_path}. Error: {e}")

# Start processing from the raw directory
process_videos(raw_base_path, processed_base_path)

print("Video processing complete.")
```
